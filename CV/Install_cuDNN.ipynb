# Install cuDNN for enable GPU used for tensorflow

## CuDNN Installation guide

https://docs.nvidia.com/deeplearning/cudnn/latest/installation/windows.html

## Steps to do

1. Update NVIDIA driver 565.90
   https://www.nvidia.com/en-us/drivers/

2. Install CUDA toolkit for windows 10 pro
   https://docs.nvidia.com/cuda/cuda-installation-guide-microsoft-windows/index.html

Note: The tensorflow=2.10 is the last version support Windows Native,
If wanna use Windows for it we need touse Python<3.11 (And ton of issue and disadvantages)
