```python
# Step 1: Import necessary packages
import pandas as pd
import os

# Step 2: Import the class from the file
from build import RegressionPipeline

# Folder path
folder_dir = 'D:/data/CaloriesPrediction/'
# Step 3: Load or define your dataframe
df = pd.read_csv(os.path.join(folder_dir, 'train.csv'))
df.info()
```

```python
df[df['Heart_Rate'] < 0]
```

```python
df = df.drop(['Sex', 'id'], axis=1, errors='ignore')
df.info()
```

```python
# Step 4: Create an instance of the pipeline
pipeline = RegressionPipeline(df=df, label='Calories', version=1)

# Step 5: Call the pipeline methods
pipeline.handle_outliers(method='zscore', threshold=3.0, cols=['Duration', 'Heart_Rate', 'Body_Temp'], note='Remove z-score outliers') \
        .scale_features(method='standard', cols=['Age', 'Height', 'Weight', 'Duration', 'Heart_Rate', 'Body_Temp'], note='Standard scaling') \
        .remove_duplicates(note='Remove duplicated rows') \
        .drop_low_variance(threshold=0.01, note='Drop near-zero variance features') \
        .train_model(test_size=0.2, random_state=42, note='Train with Linear Regression')
```
