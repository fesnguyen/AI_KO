# Choose a model
**To select the most suitable algorithm based on**:
- Problem type
- Data characteristics
- Computational resources
- Need for interpretability vs performance

## 1. Match Model to Problem Type

| Problem Type                | Example Use Case                   | Common Models                                       |
| --------------------------- | ---------------------------------- | --------------------------------------------------- |
| **Regression**              | Predict housing price              | Linear Regression, Decision Tree Regressor, XGBoost |
| **Classification**          | Spam detection, disease prediction | Logistic Regression, SVM, Random Forest             |
| **Clustering**              | Customer segmentation              | KMeans, DBSCAN, Gaussian Mixture Models             |
| **Anomaly Detection**       | Fraud detection                    | Isolation Forest, One-Class SVM, Autoencoders       |
| **Time Series Forecasting** | Stock price prediction             | ARIMA, Prophet, LSTM, XGBoost                       |
| **Recommendation**          | Product suggestions                | Matrix Factorization, LightFM, Neural CF            |


## 2. Model Selection Based on Data Size

| Dataset Size          | Suggested Models                                     |
| --------------------- | ---------------------------------------------------- |
| **Small (≤10k rows)** | Linear Models, Decision Trees, Naive Bayes           |
| **Medium (10k–100k)** | Random Forest, Gradient Boosting (XGBoost, LightGBM) |
| **Large (100k+)**     | XGBoost, LightGBM, Deep Learning (Neural Networks)   |


## 3. Start with Baseline Models

| Model                   | Task           | Strengths                        | Use When                                    |
| ----------------------- | -------------- | -------------------------------- | ------------------------------------------- |
| **Linear Regression**   | Regression     | Simple, interpretable            | Data is linear or nearly so                 |
| **Logistic Regression** | Classification | Fast, interpretable              | Binary/multiclass classification            |
| **Decision Tree**       | Both           | Intuitive, handles non-linearity | Need for interpretability or quick training |
| **DummyClassifier**     | Classification | Baseline only                    | To compare whether your model is learning   |


## 4. Advanced Models for Better Accuracy

| Model                  | Use Case               | Highlights                                            |
| ---------------------- | ---------------------- | ----------------------------------------------------- |
| **Random Forest**      | Both                   | Ensemble, handles non-linearity, robust to noise      |
| **XGBoost / LightGBM** | Both                   | High accuracy, handles missing values, fast training  |
| **CatBoost**           | Both                   | Best for categorical data, no need for much tuning    |
| **SVM**                | Classification         | Good for high-dimensional data, supports kernel trick |
| **Neural Networks**    | Any (esp. text, image) | Best for complex patterns, flexible architecture      |
| **KNN**                | Simple problems        | Very intuitive but slow for large data                |

## 6. Model Selection Tips

| Factor                | Model Preference Example                  |
| --------------------- | ----------------------------------------- |
| Small data            | Logistic/Linear Regression, Decision Tree |
| Large feature set     | SVM, Gradient Boosting                    |
| Interpretability      | Linear/Logistic Regression, Decision Tree |
| Performance focus     | XGBoost, Neural Nets                      |
| Real-time predictions | Linear Models, Tree-based                 |
| Sparse data           | Naive Bayes, Logistic Regression          |
