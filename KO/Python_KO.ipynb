## List Comprehensive

### 1. Basic List Comprehension

```python
# Create a list of numbers from 0 to 4
numbers = [x for x in range(5)]
# Output: [0, 1, 2, 3, 4]
```

### 2. List Comprehension with Condition (if statement)

```python
# Create a list of even numbers from 0 to 9
evens = [x for x in range(10) if x % 2 == 0]
# Output: [0, 2, 4, 6, 8]
```

### 3. List Comprehension with Else (if-else)

```python
# Create a list where even numbers are doubled, odd numbers are halved
modified = [x * 2 if x % 2 == 0 else x / 2 for x in range(10)]
# Output: [0, 0.5, 4, 1.5, 8, 2.5, 12, 3.5, 16, 4.5]
```

### 4. Nested Loops in List Comprehension

```python
# Create pairs of numbers from two different ranges
pairs = [(x, y) for x in range(3) for y in range(3)]
# Output: [(0, 0), (0, 1), (0, 2), (1, 0), (1, 1), (1, 2), (2, 0), (2, 1), (2, 2)]
```

### 5. List Comprehension with Multiple Conditions

```python
# List of numbers from 0 to 9 that are both even and greater than 4
filtered = [x for x in range(10) if x % 2 == 0 and x > 4]
# Output: [6, 8]
```

### 6. List Comprehension with Function Calls

```python
# Apply a function to each element in the list
def square(x):
    return x * x

squares = [square(x) for x in range(5)]
# Output: [0, 1, 4, 9, 16]
```

### 7. Flattening a List of Lists

```python
# Flatten a 2D list (list of lists) into a 1D list
matrix = [[1, 2], [3, 4], [5, 6]]
flat = [num for sublist in matrix for num in sublist]
# Output: [1, 2, 3, 4, 5, 6]
```

### 8. List Comprehension with enumerate

```python
# Create a list of (index, value) pairs from an existing list
fruits = ['apple', 'banana', 'cherry']
indexed_fruits = [(i, fruit) for i, fruit in enumerate(fruits)]
# Output: [(0, 'apple'), (1, 'banana'), (2, 'cherry')]
```

###  9. List Comprehension with zip (Pairing Two Lists)

```python
# Create pairs of elements from two lists
names = ['Alice', 'Bob', 'Charlie']
scores = [85, 90, 78]
paired = [(name, score) for name, score in zip(names, scores)]
# Output: [('Alice', 85), ('Bob', 90), ('Charlie', 78)]
```

### 10. List Comprehension with String Manipulation

```python
# Convert each word to uppercase in a list of strings
words = ["hello", "world", "python"]
uppercase_words = [word.upper() for word in words]
# Output: ['HELLO', 'WORLD', 'PYTHON']
```

### 11. List Comprehension with a Set (Remove Duplicates)

```python
# Convert a list to a set to remove duplicates, then back to a list
numbers = [1, 2, 2, 3, 3, 3]
unique_numbers = [num for num in set(numbers)]
# Output: [1, 2, 3]
```

### 12. List Comprehension with Dictionaries

```python
# Extract dictionary values using list comprehension
students = {"Alice": 90, "Bob": 85, "Charlie": 95}
scores = [score for score in students.values()]
# Output: [90, 85, 95]
```

### 13. Nested List Comprehension (Matrix Transposition)

```python
# Transpose a matrix (rows to columns and vice versa)
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
transposed = [[row[i] for row in matrix] for i in range(3)]
# Output: [[1, 4, 7], [2, 5, 8], [3, 6, 9]]
```

### 14. List Comprehension with Tuple Unpacking

```python
# Unpack tuples into separate variables within a list comprehension
pairs = [(1, 'apple'), (2, 'banana'), (3, 'cherry')]
fruits = [fruit for _, fruit in pairs]
# Output: ['apple', 'banana', 'cherry']
```

### 15. Using List Comprehension with a Generator

```python
# Using a generator to generate squares lazily
generator = (x**2 for x in range(5))  # This creates a generator, not a list

# Convert to a list
squares = list(generator)  # [0, 1, 4, 9, 16]
```

### 16. Removing None Values

```python
# Remove None values from a list
values = [1, None, 2, None, 3, 4]
cleaned = [x for x in values if x is not None]
# Output: [1, 2, 3, 4]
```

### 17. Nested List Comprehension with Conditionals

```python
# Flatten and filter a matrix of even numbers
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flattened_even = [num for row in matrix for num in row if num % 2 == 0]
# Output: [2, 4, 6, 8]
```

### 18. Working with range and Arithmetic

```python
# Generate squares of numbers from 1 to 5
squares = [x**2 for x in range(1, 6)]
# Output: [1, 4, 9, 16, 25]
```

### 19. Combining Multiple Lists

```python
# Combine three lists element-wise using list comprehension
a = [1, 2, 3]
b = [4, 5, 6]
c = [7, 8, 9]
combined = [x + y + z for x, y, z in zip(a, b, c)]
# Output: [12, 15, 18]
```

### 20. Flatten List with Filtering

```python
# Flatten a list of lists and filter out numbers greater than 5
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
filtered_flat = [num for row in matrix for num in row if num > 5]
# Output: [6, 7, 8, 9]
```
